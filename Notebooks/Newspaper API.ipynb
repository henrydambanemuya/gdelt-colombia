```python
'''
    Read the docs: http://newspaper.readthedocs.io/en/latest/
    Forked from GitHub: https://github.com/codelucas/newspaper
    Pretty Cool Demo: http://newspaper-demo.herokuapp.com/
'''

# Import useful libraries
import re
import time
import operator
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from math import isnan
from collections import Counter
from collections import OrderedDict 
from newspaper import Article
from NewspaperLanguages import translations
%matplotlib inline 
pd.set_option('display.max_colwidth', -1)

# Declare global variables
all_mentions = pd.read_csv('all_mentions.csv').sort_values('EventTimeDate', ascending=1)

filtered_languages = ['srclc:spa;eng:Moses 2.1.1 / MosesCore Europarl es-en / GT-SPA 1.0', 
                      'srclc:spa;eng:GT-SPA 1.0']

spanish_mentions = all_mentions.loc[all_mentions['MentionDocTranslationInfo'].isin(filtered_languages)]


url_list = list(spanish_mentions.MentionIdentifier)
mention_languages = list(spanish_mentions.MentionDocTranslationInfo)

authors = []
published = []
text = []
keywords = []
summaries = []
images = []
movies = []
bad_urls = 0

#article.html
```

```python
for url in url_list:    
    
    try:
        
        article = Article(url, language='es')
        #time.sleep(10)
        article.download()
        article.parse()

        authors.append(article.authors)
        published.append(article.publish_date)
        text.append(article.text.replace("\n","").strip())

        article.nlp()

        keywords.append(article.keywords)
        summaries.append(article.summary)
        images.append(article.top_image)
        movies.append(article.movies)
        
    except:
        
        authors.append('')
        published.append('')
        text.append('')
        keywords.append('')
        summaries.append(['')
        images.append('')
        movies.append('')
        
        bad_urls += 1
        
        continue
    
```

```python
spanish_mentions['Authors'] = authors
spanish_mentions['Publised'] = published
spanish_mentions['Text'] = text
spanish_mentions['Keywords'] = keywords
spanish_mentions['Summaries'] = summaries
spanish_mentions['Images'] = images
spanish_mentions['Movies'] = movies
```

```python
spanish_mentions.to_csv('event_mentions_text_es.csv', index=False)
```
